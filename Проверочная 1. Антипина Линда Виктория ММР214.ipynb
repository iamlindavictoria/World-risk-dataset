**Информация о датасете**

Датасет является ежегодным техническим отчетом о глобальных рисках бедствий. Все компоненты индекса масштабируются в диапазоне значений от 0 до 100. Разделим уровень отсутствия способности справляться на пять категорий Very High, High, Medium, Low, Very Low

```python
def get_adapt_cat(adapt):
    if adapt <= 40:
        return'Very low'
    elif adapt <= 50:
        return'Low'
    elif adapt <= 70:
        return'Medium'
    elif adapt <= 80:
        return'High'
    elif adapt <= 100:
        return'Very High'
    else:
        return adapt
```

```python
df_worldrisk['Отсутствие способности справляться_кат'] = df_worldrisk['Отсутствие способности справляться'].apply(get_adapt_cat)
```
